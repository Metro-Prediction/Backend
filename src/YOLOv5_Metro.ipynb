Counting the number of passengers detected

```python
"to count the number of passengers, take the sum of all the detected objects by accessing the labels for a particular image/timestamp"
import os

label_folder = '/content/drive/MyDrive/metro-ai-engine/results/yolo/test_run_with_labels/labels'
person_counts = {}

for label_file in os.listdir(label_folder):
    if label_file.endswith('.txt'):
        with open(os.path.join(label_folder, label_file), 'r') as f:
            lines = f.readlines()
            person_count = sum(1 for line in lines if line.startswith('0'))  # class 0 = person
            person_counts[label_file] = person_count


for image, count in person_counts.items():
    print(f"{image.replace('.txt', '.jpg')}: {count} people detected")
```
